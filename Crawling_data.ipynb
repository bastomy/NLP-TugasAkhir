# Crawling

Pada tugas ini data yang digunakan untuk dataset adalah tweet, tweet diambil dengan cara crawling menggunakan library python tweepy

terdapat dua library yang digunakan pada code ini yaitu tweepy untuk crawling data dan csv untuk eksport data menjadi data CSV. <br> untuk menggunakan tweepy kita membutuhkan beberapa key sebagai autentifikasi, key tersebut bisa diperoleh di https://developer.twitter.com/en.html <br>
berikut key yang dibutuhkan<br>
<ul>
	<li>consumer_key</li>
	<li>consumer_secret</li>
	<li>access_token</li>
	<li>access_token_secret</li>
</ul>

untuk mendapatkan key atau yang biasa disebut token diatas kita diharuskan mendaftar sebagai developer di tweeter dan mengisi beberapa form untuk verifikasi akun, setelah mendapatkan akun , token dapat digunakan menggunakan kode<br>
<ul>
	<li>self.auth = tweepy.OAuthHandler(self.consumer_key, self.consumer_secret)</li>
	<li>self.auth.set_access_token(self.access_token, self.access_token_secret)</li>
</ul>

<ul>
	<li>id tweet</li>
	<li>id user</li>

```python
import tweepy
import csv

class Services:

    def __init__(self):
        self.consumer_key = 'aqh26LmWvnCwEqoWnBkNDgq3X'
        self.consumer_secret = 'WLTGI57tcvei5AwtZNxiyrqDmTqhmVEaeQlXhknzIFZBym58Fv'
        self.access_token = '232106297-SLMzoxQqdrOboZlOs8VLShayem1FUr99pGE8Y8Wi'
        self.access_token_secret = '8qlseHuNKDsnIVnxuxDZcXfjDepCrWDcp5U1JtZ4PfOU9'
        self.auth = tweepy.OAuthHandler(self.consumer_key, self.consumer_secret)
        self.auth.set_access_token(self.access_token, self.access_token_secret)
        self.api = tweepy.API(self.auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

    def getData(self,parameter):
        print("starting")
        i = 0
        with open('data_crawling.csv', 'w', newline='') as csvFile: # membuat file csv dan membukanya
            writer = csv.writer(csvFile) #membuat csv writter
            row = ["ID Tweet", "ID User", "Screen Name", "Tweet", "Timestamp"] #membuat header pada data crawling
            writer.writerow(row) # menulis data header
            for post in tweepy.Cursor(self.api.search, q=parameter, tweet_mode='extended').items(): 
                # tweepy menggunakan cursor untuk mencari sesuai parameter yaitu @prabowo
                # tweet_mode="extended" digunakan untuk mendapatkan full text dari tweet 
                try:
                    if (post.retweet_count == 0): #menghindari tweet hasil retweet
                        if i <= 99 : #membatasi hanya 100 data saja yang akan diambil
                            row = [post.id_str.encode('utf-8'), post.user.id_str.encode('utf-8'), post.user.screen_name,
                                   post.full_text.encode('utf-8'), post.created_at] 
                            # data seperti id_tweet,id_user di encode karena data integer terlalu besar yang tidak bisa d handle 
                            # oleh csv sehingga harus d encode agar mendapatkan full number integer
                            # untuk full_text (tweet) di encode untuk menghindari dari emoticon yang menyebabkan error saat di tulis ke csv
                            writer.writerow(row) #perintah untuk menulis ke csv
                        else:
                            break
                        i = i + 1 # iterasi
                        print(i, " tweet downloaded ", post.created_at)
                except tweepy.RateLimitError: # untuk menghandle apabila token telah mencapai batas maksimum data yang diperbolehkan
                    print("sleep") # crawling akan di jeda sampai token mendapat resource untuk crawling kembali
                except tweepy.TweepError as e: # untuk menhandle apabila terjadi kesalahan saat crawling
                    print('tweepy error')
                    print(e.reason)
                    continue
                except StopIteration:  # stop iteration ketika tidak ada data yang tersedia lagi
                    print('habis')
                    break
        csvFile.close() #close file csv
        return True
```

```python
search = "@prabowo"
if (Services().getData(search)):
    print("successfully")
else:
    print("failed")
```
